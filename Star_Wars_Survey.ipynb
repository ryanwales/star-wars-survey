## Project Goal

Does America realize that "The Empire Strikes Back" is the best Star Wars movie?

Let's answer this question using the [this dataset](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")

star_wars.head(10)
```

```python
star_wars.columns
```

There are two columns that are Yes/No questions. They are currently coded as strings, but we want to change them to Bools because they're easier to work with.

Here are the columns:
- `Have you seen any of the 6 films in the Star Wars franchise?`
- `Do you consider yourself to be a fan of the Star Wars film franchise?`

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

```python
# mapping dictionary
yes_no = {'Yes': True, 'No': False}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)

star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = \
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)

# check out values
star_wars.head()
```

Similarly, we have some columns that ask which Star Wars film that a person has seen. These data are represented as checkboxes, where a value is either the name of a movie or `NaN` if the person didn't see the movie. Let's change these to be Booleans.

```python
# First, convert the values to Boolean using the same map technique
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)

star_wars.head()
```

Now we can rename the columns to something more intuitive.

```python
star_wars.rename(columns={
'Which of the following Star Wars films have you seen? Please select all that apply.':'seen_1',
'Unnamed: 4':'seen_2',
'Unnamed: 5':'seen_3',
'Unnamed: 6':'seen_4',
'Unnamed: 7':'seen_5',
'Unnamed: 8':'seen_6'
    },inplace=True
)

star_wars.head()
```

### Clean the Ranking Columns

Participants were asked to rank the Star Wars movies from 1-6, with 1 being their favorite.

Let's convert all of the columns corresponding to the these rankings to `float`.

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

# rename ranking columns
for col,i in zip(star_wars.columns[9:15],range(6)):
    new_col = 'ranking_' + str(i+1)
    star_wars.rename(columns={col:new_col},inplace=True)
    
star_wars.columns
```

Now that we've cleaned these columns, we can now perform operations on them. We can also begin to visualize our data.

### View count

```python
# Calculate the sum of each of the seen columns
ep1_sum = star_wars['seen_1'].sum()
ep2_sum = star_wars['seen_2'].sum()
ep3_sum = star_wars['seen_3'].sum()
ep4_sum = star_wars['seen_4'].sum()
ep5_sum = star_wars['seen_5'].sum()
ep6_sum = star_wars['seen_6'].sum()

print('Episode 1:',ep1_sum,'\n',
      'Episode 2:',ep2_sum,'\n',
      'Episode 3:',ep3_sum,'\n',
      'Episode 4:',ep4_sum,'\n',
      'Episode 5:',ep5_sum,'\n',
      'Episode 6:',ep6_sum,'\n',)
```

Episodes 5 and 6 were much more viewed than other episodes. Episodes 2 and 3 were the least viewed episodes.

### Ranking

```python
# Plot ranking averages
import matplotlib.pyplot as plt
%matplotlib inline

seen_cols = star_wars.columns[9:15]
plt.bar(range(6),star_wars[seen_cols].mean())
```

These are ranking values, so a lower ranking means that viewers preferred it. Episode 5 has the best rank compared to the other episodes, while Episode 3 has the worst rank. 

However, Episode 5 also had the most views, while Episode 3 had the fewest views. Therefore, views must be considered when interpreting ranking.

### Exploring the Data by Binary Segments

Let's see if the results of our analysis vary if we group people by whether they classify themselves as Star Wars fans or not. 



```python
fans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']==True]

#plot the rankings
plt.bar(range(6),fans[seen_cols].mean())
```

```python
# See how non-fans compare
nonfans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']==False]

#plot the rankings
plt.bar(range(6),nonfans[seen_cols].mean())
```

Fans and non-fans of Star Wars have dissimilar views of the prequels (Episodes 1-3), but their views of the original series (Episodes 4-6) are similar. Both groups rank Episode 4: A New Hope as the best Star Wars film.
